#Maquinas de vectores de soporte (SVM)
En comparación a otros clasificadores, ofrece una precisión muy alta. Se puede utilziar en detección de rostros de intrusos, clasificación de correos electrónicos, articulos de noticias y páginas web, etc.

Genera un hiperplano óptimo de forma iterativa, este es el hiperplano marginal máximo que mejor divide el conjunto de datos en clases. ![texto alternativo](https://miro.medium.com/max/544/1*6U9NrruycDBsPOyivpn8UQ.png) 

Kernel: transforma un espacio de datos en la forma requerida. 

Ventajas:
*   Buena precisión y más rápido que Naive Bayes.
*   Usa menos memoria porque utiliza un suconjunto de puntos de entrenamiento en la fase de desición.

Desventajas:
*   No son adecuadas para grandes conjuntos de información, ya que requiere grandes tiempos de formación.
*   Funciona mal con clases superpuestas y también es sensible al tipo de núcleo utilizado.

# Kernel
Poseen funciones núcleo, que les permiten operar en un espacio ímplicito y de alta dimensión sin tener que calcular nunca las coordenadas de los datos en ese espacio, sino que simplemente calculando los productos internos entre las imágenes de todos los pares de datos en el espacio de características. En resumen: mapea los datos desde un espacio bidimensional a un espacio tridimensional, esto permite que los datos se dividan más claramente para tomar decisiones. 
![texto alternativo](https://miro.medium.com/max/838/0*5AtpuJ7X08V4zc9f.png)
 



```
from sklearn import datasets

data = datasets.load_breast_cancer()

X = data.data
y = data.target

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.svm import SVC

svm = SVC(kernel='linear')

svm.fit(X_train, y_train)

y_pred = svm.predict(X_test)

from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```
from sklearn.metrics import precision_score

print("precision: ", precision_score(y_test, y_pred))
```

```
from sklearn.metrics import f1_score

print("f1: ", f1_score(y_test, y_pred))
```
