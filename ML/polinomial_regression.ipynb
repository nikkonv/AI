```
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

boston = datasets.load_boston()

boston.data.shape
```

```
print("columnas: ", boston.feature_names)
```

```
X_p = boston.data[:, np.newaxis, 5]
y_p = boston.target

plt.scatter(X_p,y_p)
plt.show()
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_p, y_p, test_size=0.2)
```

```
from sklearn.preprocessing import PolynomialFeatures

poli_reg = PolynomialFeatures(degree=2)

# caracteristicas existentes en caracteristicas de mayor grado
X_train_poli = poli_reg.fit_transform(X_train)
X_test_poli = poli_reg.fit_transform(X_test)

# algoritmo
pr = linear_model.LinearRegression()

# entrenar
pr.fit(X_train_poli, y_train)

Y_pred_pr = pr.predict(X_test_poli)
```

```
plt.scatter(X_test, y_test)
plt.plot(X_test, Y_pred_pr, color="red", linewidth=2)
plt.show()
```

```
print("pendiente: ", pr.coef_)
print("interseccion: ", pr.intercept_)
print("precision: ",pr.score(X_train_poli, y_train))
```

```
### other example

xe = np.array([0,1,2,3,4,5])
ye = np.array([0,0.8,0.9,0.1,-0.8,-1])

# linear
p1 = np.polyfit(xe,ye,1)
# quadratic
p2 = np.polyfit(xe,ye,2)

# coefs 
print(p1)
print(p2)
```

```
import matplotlib.pyplot as plt

plt.plot(xe,ye,'o')
plt.plot(xe, np.polyval(p1,xe), 'r-')
plt.plot(xe, np.polyval(p2,xe), 'g:')
plt.show()
```
