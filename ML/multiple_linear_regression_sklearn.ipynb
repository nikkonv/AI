```
from sklearn import datasets, linear_model
boston = datasets.load_boston()
boston.keys()
```

```
boston.data.shape
```

```
boston.feature_names
```

```
# definir datos
X_mul = boston.data[:, 5:8]
y_mul = boston.target
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_mul, y_mul, test_size=0.2)

lr_mul = linear_model.LinearRegression()

lr_mul.fit(X_train, y_train)
```

```
Y_pred_mul = lr_mul.predict(X_test)

print("pendientes: ",lr_mul.coef_)
print("interseccion: ", lr_mul.intercept_)
print("precision: ",lr_mul.score(X_train, y_train))
```
