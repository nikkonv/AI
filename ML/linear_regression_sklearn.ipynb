```
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
```

```
boston = datasets.load_boston()
boston.keys()
```

```
print(boston.DESCR)
```

```
# rm numero de habitaciones 
print(boston.feature_names)
```

```
# rm
X = boston.data[:,np.newaxis,5]
# etiquetas
y = boston.target
```

```
plt.scatter(X,y)
plt.xlabel('numero de habitaciones')
plt.ylabel('valor medio')
plt.show()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

lr = linear_model.LinearRegression()

lr.fit(X_train, y_train)

Y_pred = lr.predict(X_test)

plt.scatter(X_test, y_test)
plt.plot(X_test, Y_pred, color="red", linewidth = 3)
plt.title('regresion lineal simple')
plt.xlabel('numero de habitaciones')
plt.ylabel('valor medio')
plt.show()
```

```
# ecuacion del modelo
print("pendiente o coeficiente 'a': ",lr.coef_)
print("interseccion o coeficiente 'b': ", lr.intercept_)
print("y = {}x + {}".format(float(lr.coef_), float(lr.intercept_)))
```

```
# precision r^2
print("percision: ", lr.score(X_train, y_train))
```
