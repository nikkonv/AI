# Agrupamiento por K-medios

Es uno de los algoritmos no supervisado más simples y populares en machine learning. Se encarga de agrupar datos segun similitudes en sus caracteristicas.

En otras palabras, el algoritmo K-means identifica k número de centroides, y luego asigna cada punto de datos al grupo más cercano, mientras mantiene los centroides lo más pequeños posible.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
path = '/content/drive/My Drive/Colab Notebooks/ML/resources/movies.csv'
data = pd.read_csv(path)
data = pd.DataFrame(data)
data.columns
```

```
x = data[data.columns[0]].values
y = data[data.columns[1]].values
print(x,y)
```

```
a = [1,2,3]
b= [4,5,6]
list(zip(a,b))
```

```
X = np.array(list(zip(x, y)))
print(X[:3])
```

```
km = KMeans(n_clusters=5)
km.fit(X)
labels = km.predict(X)
centroids = km.cluster_centers_
```

```
r = np.copy(X)
labels = [[i] for i in labels]

r = np.append(r, np.copy(labels), axis=1)
```

```
np.amax(r)
```

```
colors = ['r.','m.','c.','y.','b.']
plt.scatter(centroids[:,0], centroids[:,1], marker='x', s=150, linewidths=5, zorder=10)

for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)
plt.show()
```
