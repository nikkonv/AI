```
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets

boston = datasets.load_boston()
```

```
X_adr = boston.data[:, np.newaxis, 5]

y_adr = boston.target

plt.scatter(X_adr, y_adr)
plt.show()
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_adr, y_adr, test_size=0.2)

from sklearn.tree import DecisionTreeRegressor

adr = DecisionTreeRegressor(max_depth=6)

adr.fit(X_train, y_train)

Y_pred = adr.predict(X_test)

X_grid = np.arange(min(X_test), max(X_test), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X_test, y_test)
plt.plot(X_grid, adr.predict(X_grid), color="red", linewidth=2)
plt.show()
```

```
print("precision: ", adr.score(X_train, y_train))
```

Al cambiar "max_depth" en la definici√≥n del modelo, incrementa el score del mismo, sin embargo, llega un punto en donde se consigue un sobreajuste del modelo, y esto es negativo.
