# #3 Gradient descent 

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt 
```

Optimizar funciones (cualquiera que sea derivable obvious)

```python
#funcion anonima th: vector de parametos
#th[0] = x, th[1] = y

func = lambda th: np.sin(1/2*th[0]**2 - 1/4*th[1]**2+3) * np.cos(2*th[0]+1 - np.e**th[1])

# vista de pajaro (desde arriba)

res = 100 #resolucion

_X = np.linspace(-2,2,res)
_Y = np.linspace(-2,2,res)

_Z = np.zeros((res,res))

for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy,ix] = func([x,y])
    
plt.contourf(_X,_Y,_Z,100)
plt.colorbar()
theta = np.random.rand(2)*4/-2

_T = np.copy(theta)
h = 0.001
lr = 0.01 # learning rate

plt.plot(theta[0],theta[1],"o",C="white")

grad = np.zeros(2)

for _ in range(10000):
    for it,th in enumerate(theta):
        _T = np.copy(theta)
        _T[it] = _T[it] + h

        deriv = (func(_T) - func(theta))/h

        grad[it] = deriv
    
    theta = theta - lr * grad
    
    if (_ % 100 == 0):
        plt.plot(theta[0],theta[1],"o",C="red")

plt.plot(theta[0],theta[1],"o",C="green")
plt.show()

# ir tanteando el terreno y para donde este la pendiente hay que moverse al contrario
# poco a poco se llegara a un valle 

```
